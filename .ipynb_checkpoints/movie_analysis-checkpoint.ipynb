```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in movies.csv as a DataFrame
df = pd.read_csv('movies.csv')

# Preview the first 5 rows of the DataFrame
print(df.head())
```

## What was the runtime for each movie?

```python
df.loc[df['Movie Title'] == "Birdman or (The Unexpected Virtue of Ignorance)", 'Movie Title'] = "Birdman"
```

```python
print(df[['Movie Title', 'Runtime']])
```

```python
median_runtime = df['Runtime'].median()
print(f"The median runtime of the movies is {median_runtime} minutes.")
```

```python
plt.bar(df['Movie Title'], df['Runtime'])
plt.xticks(rotation=90)
plt.xlabel('Movie Title', color='red')
plt.ylabel('Runtime (min)', color='red')
plt.title('Runtime for Each Movie')

# Save the figure as a PNG file
plt.savefig('runtime_bar.png', dpi=300, bbox_inches='tight')

# Display the figure
plt.show()
```

## What movie had the longest runtime?


```python
longest_runtime = df['Runtime'].max()
longest_movie = df.loc[df['Runtime'] == longest_runtime, 'Movie Title'].iloc[0]
print(f"The movie with the longest runtime is {longest_movie}, with a runtime of {longest_runtime} minutes.")
```

```python
import matplotlib.pyplot as plt

# Create a box plot of movie runtimes
plt.boxplot(df['Runtime'])

# Get the maximum runtime and the movie title
longest_runtime = df['Runtime'].max()
longest_movie = df.loc[df['Runtime'] == longest_runtime, 'Movie Title'].iloc[0]

# Add a red dot for the longest runtime
plt.plot(1, longest_runtime, marker='o', markersize=10, color='red', label=f'Longest Movie: {longest_movie}')

# Set the title and axis labels
plt.title('Distribution of Movie Runtimes')
plt.ylabel('Runtime (minutes)')

# Add a legend in the bottom left corner
plt.legend(loc='lower left')

# Save the plot as a PDF file
plt.savefig('runtime_boxplot3.png', bbox_inches='tight')

# Show the plot
plt.show()
```

## Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?

```python
df_subset = df.loc[:, ['Runtime', 'Award Wins']]
print(df_subset)
```

```python


# Create a scatter plot of movie runtimes vs. award wins
axis1 = df.plot.scatter(y='Runtime', x='Award Wins', c='DarkBlue')

# Set the axis labels and title
axis1.set_xlabel('Award Wins')
axis1.set_ylabel('Runtime (min)')
axis1.set_title('Relationship between Runtime and Award Wins')

# Save the plot as a PNG file
plt.savefig('runtime_award_scatter.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

## (Graph) How many awards and nominations did each movie receive?

```python
import matplotlib.pyplot as plt

# Create a new DataFrame with only the relevant columns
awards_df = df[['Movie Title', 'Award Wins', 'Award Nominations']]

# Set the Movie Title column as the index
awards_df.set_index('Movie Title', inplace=True)

# Plot a horizontal bar chart
awards_df.plot(kind='barh', stacked=True)

# Set the axis labels and title
plt.xlabel('Number of Awards')
plt.ylabel('Movie Title')
plt.title('Number of Awards and Nominations for Each Movie')

# Save the plot as a PNG file
plt.savefig('awards_barh.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
